tryiing to do this in a loop

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,20))
for r in range(1,8):
    ax = fig.add_subplot(8,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist')
    ax.set_title(cols[r])
```
